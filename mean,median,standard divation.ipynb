```python
#WAP for standared deviation
x = [1,2,3,4,5]
sm = 0
mean = sum(x)/len(x)
for i in range(len(x)):
    sm=sm+(x[i]-mean)**2
rmse=(sm/len(x))**0.5
print(mean)
print(rmse)
```

```python
#write a program for find mean
x = [1,2,3,4,5]
s=0
for i in range(len(x)):
    s=s+x[i]
mean=s/len(x)
print(mean)
```

```python
#write a program for find median
x = [1,2,3,5,4]
x.sort()
n=len(x)
if n%2==0:
    mid1=x[n//2]
    mid2=x[n//2-1]
    median=(mid1+mid2)/2
else:
    median=x[n//2]
print(median)

```
